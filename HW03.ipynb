Next we will find the index of coincidence for each decimation (every 3 letters shold have an index of coincidence close to the Egnlish language 0.065) 


```python
import collections
from collections import Counter

def IC(s):
    f = dict()
    f = dict(Counter(''.join(s)))
    od = collections.OrderedDict(sorted(f.items()))
    
    sum_IC = 0
    N = len(s)
    
    for key in od:
        c = od[key]
        sum_IC += ( (c*(c-1)) / (N*(N-1)) )
        
    return round(sum_IC, 3)

tot = 0
for i in range(3):
    print("The index of coincidence for decimation", i+1, "is: ", IC(l[i]))
    tot+=IC(l[i])
print("The average IC is: ",tot/3)
```

Our average index of coincidence gives us additional evidence that the period is 3 because the index of coincidence is very close to the index of coincidence for regular english is around 0.06
